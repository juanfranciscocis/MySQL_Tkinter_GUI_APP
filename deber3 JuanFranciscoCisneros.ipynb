```python
import tkinter as tk
import mysql
from mysql import connector



class DB:
    def __init__(self, host, user, password, database):
        self.connection = mysql.connector.connect(
            host=host,
            user=user,
            password=password,
            database=database
        )
        self.cursor = self.connection.cursor()
        self.is_successful = False

    def execute_query(self, query):
        self.is_successful = False
        try:
            self.cursor.execute(query)
            self.connection.commit()
            print("Query executed successfully")
            self.is_successful = True
        except mysql.connector.Error as e:
            print(f"The error '{e}' occurred")
        return self.is_successful

    def execute_read_query(self, query):
        self.is_successful = False
        try:
            self.cursor.execute(query)
            result = self.cursor.fetchall()
            for row in result:
                print(row)
            self.is_successful = True
            return result
        except mysql.connector.Error as err:
            print(f"Error: {err}")
            return None
    
def create_table(header, data):
    # Create table header
    for i in range(len(header)):
        tk.Label(table, text=header[i], relief=tk.RIDGE, width=25).grid(row=0, column=i)

    # loop through data
    for i in range(len(data)):
        for j in range(len(data[i])):
            tk.Label(table, text=data[i][j], relief=tk.RIDGE, width=25).grid(row=i+1, column=j)


def update_success_label(success):
    if success:
        success_label.pack()
        fail_label.pack_forget()
    else:
        fail_label.pack()
        success_label.pack_forget()


root = tk.Tk() 
# root window title and dimension
root.title("Deber 3 - Base de Datos")
# Set geometry (widthxheight)
root.geometry('900x900')


# connect to database
db = DB("localhost", "root", "06012002jF_", "RunningApp")


table = tk.Frame(root)
table.pack()


# create a label for the header
header_label = tk.Label(root, text="Header (separate by comma)")
header_label.pack()
# create a input field for the header
header = tk.Entry(root, width=50)
header.pack()

# create a label for the query
query_label = tk.Label(root, text="Query To Show (SQL syntax of a SELECT)")
query_label.pack()
# create a input field for the query
query = tk.Entry(root, width=50)
query.pack()


# create a submit button
submit = tk.Button(root, text="Submit Select (Will Show the table)", command=lambda: (create_table(header.get().split(","), db.execute_read_query(query.get())), update_success_label(db.is_successful)))
submit.pack()

# create a label for the query
query_label = tk.Label(root, text="Query To Execute (SQL syntax of a INSERT, UPDATE, DELETE or OTHERS)")
query_label.pack()
# create a input field for the query
query_2 = tk.Entry(root, width=50)
query_2.pack()
# create a submit button
submit_two = tk.Button(root, text="Submit Query", command=lambda: (db.execute_query(query_2.get()), update_success_label(db.is_successful)))
submit_two.pack()


# if the query is successful
success_label = tk.Label(root, text="Query Successful", fg="green")
# if the query is not successful
fail_label = tk.Label(root, text="Query Failed", fg="red")


# init the labels as hidden
success_label.pack_forget()
fail_label.pack_forget()
    




# all widgets will be here
# Execute Tkinter
root.mainloop()


```
